```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0, 256, 1)

gamma = []

for i in [0.1, 0.3, 0.6, 1, 2, 3.5, 10]:
    g = (a / 255) ** i * 255
    gamma.append(g)

plt.figure(figsize=(10, 6))
plt.plot(a, gamma[0], label='Gamma = 0.1')
plt.plot(a, gamma[1], label='Gamma = 0.3')
plt.plot(a, gamma[2], label='Gamma = 0.6')
plt.plot(a, gamma[3], label='Gamma = 1', linewidth=3 , color = "black")  # Making the line thicker for gamma = 1
plt.plot(a, gamma[4], label='Gamma = 2')
plt.plot(a, gamma[5], label='Gamma = 3.5')
plt.plot(a, gamma[6], label='Gamma = 10')

plt.title('Gamma Correction Curves')
plt.xlabel('Input Intensity (a)')
plt.ylabel('Output Intensity (gamma)')

plt.grid(True)
plt.legend(fontsize='small') 
plt.show()
```
